```python
import unirna
from transformers import AutoTokenizer, AutoModel

source_path = "../weights/unirna_L16_E1024_DPRNA500M_STEP400K"
tokenizer = AutoTokenizer.from_pretrained(source_path)
model = AutoModel.from_pretrained(source_path)
```

```python
seq = "ACACTGGGNNNNUUUU"
token = tokenizer(seq, return_tensors="pt")
model(**token, output_attentions=True)
```
